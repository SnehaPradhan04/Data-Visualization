```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x, y, c='g', alpha=0.5, marker='X')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x vs y plot")
plt.grid()
```

```python
x = np.linspace(1, 10, 30)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.plot(x, y, '--r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sinusoidal curve')
plt.grid()
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,5))
plt.bar(x, y, color='y')
plt.xlabel('categorical data')
plt.ylabel('numerical data')
plt.title('bar plot')
plt.grid()
```

```python
plt.figure(figsize=(5,5))
plt.barh(x, y, color='m')
plt.xlabel('numerical data')
plt.ylabel('categorical data')
plt.title('bar plot')
plt.grid()
```

```python
x = [3,4,5,6,7,8]
y = [4,5,6,7,8,9]

plt.figure(figsize=(4,2))
plt.plot(x,y)
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x,y)
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8]
plt.figure(figsize=(6,4))
plt.hist(data, color='#C8AEE7')
plt.xlabel('unique values')
plt.ylabel('frequency')
plt.title('histogram plot')
plt.grid()
plt.savefig("histogram.png")
```

```python
x = np.random.rand(30)
y = np.random.rand(30)
z = np.random.rand(30)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c='y',alpha=.8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
