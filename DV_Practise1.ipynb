```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

weather_data = {
    'Ravangla_High': [21, 16, 20, 19, 15, 16, 18, 19, 20, 15],
    'Ravangla_Low': [2, -1, 3, 6, 4, -2, 1, 5, 3, 1],
    'Bbsr_High': [45, 40, 45, 41, 39, 39, 40, 46, 42, 42],
    'Bbsr_Low': [19, 20, 23, 26, 27, 25, 18, 19, 20, 21]
}
df = pd.DataFrame(weather_data)

df
```

```python
plt.plot(df)
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('simple plot')
plt.legend(['Ravangla_high','Ravangla_low','Bbsr_high','Bbsr_low'])
plt.grid()
```

```python
sns.scatterplot(df)
plt.title('Scatter Plot')
plt.xlabel('days')
plt.ylabel('temperature')
plt.legend(['Ravangla_high','Ravangla_low','Bbsr_high','Bbsr_low'])
plt.grid()
plt.show()
```
